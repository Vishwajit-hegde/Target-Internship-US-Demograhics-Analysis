```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Data_Profiles.csv',index_col='zip code tabulation area')
```

```python
all_cols = list(df.columns)
percent_cols = []
for col in all_cols:
    if(col[-2:]=='PE'):
        percent_cols.append(col)
```

```python
percent_df = df[percent_cols]
```

```python
PR_zipcodes = percent_df[percent_df['DP02_0001PE'].isna()].index
PR_attributes = [x for x in percent_cols if x[4:6]=='PR']
nonPR_attributes = [x for x in percent_cols if x[4:6]!='PR']
for zipcode in PR_zipcodes:
    for attr1,attr2 in zip(PR_attributes,nonPR_attributes):
        percent_df.loc[zipcode,attr2] = percent_df.loc[zipcode,attr1]
for attr in PR_attributes:
    percent_df.drop(attr,axis=1,inplace=True)
```

```python
percent_df.dropna()
```

```python
def r(a,b):
    return list(range(a,b+1))
DP02_numbers = r(1,36)+[43,44,50,51]+r(52,60)+r(61,71)+r(78,86)+[92]+r(110,121)+r(123,149)+[151,152]
DP03_numbers = r(1,17)+r(26,31)+r(33,45)+r(47,50)+r(52,63)+r(96,101)+[119]
DP05_numbers = r(2,32)+r(36,62)+r(64,69)+[71]+r(87,89)
imp_attr_list = []
c = 0
for j in [DP02_numbers,DP03_numbers,DP05_numbers]:
    if(c==0):
        table = 'DP02_'
    elif(c==1):
        table = 'DP03_'
    else:
        table = 'DP05_'
    for i in j:
        if(i<10):
            attr = table + '000' + str(i) + 'PE'
        elif(i<100):
            attr = table + '00' + str(i) + 'PE'
        else:
            attr = table + '0' + str(i) + 'PE'
        imp_attr_list.append(attr)
    c += 1
```

```python
imp_df = percent_df[imp_attr_list]
```

```python
for col in imp_df.columns[(imp_df<0).all()]:
    col_ = col[:9]+col[-1:]
    imp_df[col] = df[col_]
```

```python
for col in imp_df.columns:
    imp_df.loc[imp_df[col]<0,col] = np.nan
```

```python
imp_df = imp_df.fillna(imp_df.median())
```

```python
imp_df.to_csv('Cleaned_percent_data.csv')
```
